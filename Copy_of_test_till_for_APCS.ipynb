**Exercise 1: Vowel Counter** (5 minutes)

Write a Python program that:

1. Asks the user to input a word.
2. Counts the number of vowels (a, e, i, o, u) in that word.
3. Prints "The word has X vowels!", replacing X with the number of vowels.

For example, if the user enters "hello", the program should print "The word has 2 vowels!".

````python
#Ex1
word = input("enter a word: ")
vowels = "aeiou"
x=0

for i in range(0, len(word), 1):
  if word in vowels:
    x+=1

print("the word has ", x, "vowels")
````

**Exercise 2: Even Number Finder** (7 minutes)

Write a Python program that:

1. Takes a list of numbers: `[12, 34, 55, 66, 73, 86, 99]`.
2. Iterates over the list to find all even numbers.
3. Prints out each even number from the list.

For instance, based on the given list, the program should print:
```
12
34
66
86
```



````python
#Ex2
x = [12, 34, 55, 66, 73, 86, 99]

for i in range(0, len(x), 1):
  if x[i] % 2 == 0:
    print(x[i])
````

**Exercise 3: Accumulator Pattern** (7 minutes)

Given a set limit, create a program that:

1. Starts with a number `0`.
2. Continuously adds `5` to the number using a `while` loop.
3. Stops the loop once the number exceeds the given limit.
4. Prints the last number before exceeding the limit.

For instance, if the limit is `23`, the program should print `20` as adding another `5` would exceed the limit.

````python
#Ex3
num = 0
limit = 23


while num < limit:
  print(num)
  num += 5
````

**Exercise 4: Multiplication Table Generator** (7 minutes)

Write a Python program that:

1. Prompts the user to input a number.
2. Displays the multiplication table for that number up to 10.
3. Each line should be formatted as: `N x M = P`, where `N` is the input number, `M` is the multiplier (from 1 to 10), and `P` is the product.

For example, if the user enters `5`, the program should print:
```
5 x 1 = 5
5 x 2 = 10
5 x 3 = 15
...
5 x 10 = 50
```



````python
#Ex4
n = int(input("enter a number: "))

for m in range (0, 11, 1):
  print(n, "*", m, "=", int(n*m))
````

**Exercise 5: Number Guessing Game** (8 minutes)

Implement a simple number guessing game:

1. Choose a random number between 1 and 10 (you can set this number manually for this exercise).
2. Prompt the user to guess the number.
3. Use a `while` loop to allow the user to keep guessing until they guess the correct number.
4. After each guess, inform the user if their guess was too high, too low, or correct.
5. Once the user guesses correctly, print "Congratulations! You guessed the right number."

For example, if the chosen number is `7` and the user first guesses `5`, the program should print "Too low! Try again.".

---

````python
#Ex5
ran_num = 5
guess = int(input("guess a number between 1-10: "))
while guess != ran_num:
  guess1 = int(input("Try again: "))
  if guess1 > ran_num:
    print("too high")
  elif guess1 < ran_num:
    print("too low")
  else:
    print("correct answer")
    break
````
